```python
import pandas as pd

df = pd.read_excel('NLP_Quotes.xlsx')
df = df.sample(frac=1).reset_index(drop=True)
df_bins = pd.qcut(df.index, 4, labels=False)
```

```python
df.iloc[df_bins == 0].to_excel("Rashmi.xlsx")
df.iloc[df_bins == 1].to_excel("Rohit.xlsx")
df.iloc[df_bins == 2].to_excel("Vaishnavi.xlsx")
df.iloc[df_bins == 3].to_excel("Varun.xlsx")
```
