Things to do
1. Remove null values
2. Convert all quotes to lower case
3. Check for duplicate quotes
4. Check most frequent tags
5. Filter dataset based on top-k frequent tags

```python
dataset = pd.read_csv('quotes.csv', delimiter=',')
nRow, nCol = dataset.shape
print(f'There are {nRow} rows and {nCol} columns')
```

```python
# Null Value check
print(dataset.isnull().sum())
dataset.dropna(inplace=True)
dataset.quote.drop_duplicates(inplace=True)

# Remove commas from categories. ['Life, Motivation']
dataset['category'] = dataset['category'].str.replace(', ', ' ')

# Strip extra spaces in category
dataset['category'] = dataset['category'].str.strip()
dataset['quote'] = dataset['quote'].str.strip()
dataset['author'] = dataset['author'].str.strip()

# Convert upper case to lower case
dataset['category'] = dataset['category'].str.lower()
dataset['quote'] = dataset['quote'].str.lower() ## comment if needed
dataset['author'] = dataset['author'].str.lower() ## comment if needed
```

```python
dataset.head()
```

```python
# Creating a dictionary to store count for all tags.

tags_dict = {}

for row in dataset.itertuples():
    for cat in row.category.split(' '):
        if cat=='':
            continue
        tags_dict[cat.lower()] = tags_dict.get(cat.lower(), 0) + 1
        
sorted_tags_dict = dict(sorted(tags_dict.items(), key=lambda x: x[1], reverse=True))
```

```python
print(f'There are {len(sorted_tags_dict)} types of different tags')
```

```python
sorted_tags_dict
```

```python
# Remove all the tags whose count is below the specified threshold

threshold = 1500

tags_to_remove = [key for key, value in sorted_tags_dict.items() if value < threshold]

# Remove items from the dictionary
for key in tags_to_remove:
    del sorted_tags_dict[key]
    
print(f'After removing less frequent tags: there are {len(sorted_tags_dict)} types of different tags')

# Null Value check
dataset.dropna(inplace=True)
```

```python
sorted_tags_dict
```

```python
# List of tags to keep
top_k = 5
tags_to_keep = [tag for tag in list(sorted_tags_dict.keys())[:top_k]]

def process_tags(x):
    unique_tags = list(set(tag for tag in x.split() if tag in tags_to_keep))
    
    if len(unique_tags) != 0:
        return unique_tags
    
    return None


# dataset['tags'] = dataset['category'].apply(lambda x: ' '.join(tag for tag in x.split() if tag in tags_to_keep))
dataset['tags'] = dataset['category'].apply(lambda x: process_tags(x))

## Remove quotes which are not related to any of the chosen tags
dataset.dropna(inplace=True)

## Create boolean columns similar to one hot encoding.
# tags_dummies = df['tags'].str.get_dummies(sep=' ')
# df = pd.concat([df, tags_dummies], axis=1)
# df.drop('tags', axis=1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
import re

def is_english(text):
    try:
        return detect(text) == 'en'
    except:
        return False
    
def remove_special_characters(text):
    # Define pattern for special characters
    pattern = r'[^a-zA-Z0-9\s]'  # Keep only letters, numbers, and whitespace
    
    # Remove special characters using regular expression
    text = re.sub(pattern, '', text)
    
    return text

def clean_abstract(text):
    text = text.split()
    text = [x.strip() for x in text]
    text = [x.replace('\n', ' ').replace('\t', ' ') for x in text]
    text = ' '.join(text)
    text = re.sub('([.,!?()])', r' \1 ', text)
    return text

train_proportion = 0.8  # 80% for training
test_proportion = 1 - train_proportion  # Remaining for testing
df = dataset.sample(frac=0.1)


df['quote'] = df['quote'].apply(remove_special_characters)

df['is_english'] = df['quote'].apply(is_english)
print(df['is_english'].value_counts())
print(df[df['is_english'] == False])
df = df[df['is_english']]
df = df.drop(columns = ['is_english'])

# Shuffle the DataFrame to ensure randomness
df_shuffled = df.sample(frac=1, random_state=42)

# Calculate the number of samples for training and testing
train_size = int(train_proportion * len(df_shuffled))
test_size = len(df_shuffled) - train_size

# Split the shuffled DataFrame into training and testing sets
train_set = df_shuffled.iloc[:train_size]
test_set = df_shuffled.iloc[train_size:]
train_set.to_csv('train.csv', index=False)
test_set.to_csv('test.csv', index=False)
```
